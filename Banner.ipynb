```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os datasets
df_Rnorte = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/TRABALHO/historico_regiao_norte.csv')
df_Rnordeste = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/TRABALHO/historico_regiao_nordeste1.csv')
df_Rsul = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/TRABALHO/historico_regiao_sul.csv')
df_RcentroOeste = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/TRABALHO/historico_regiao_centro-oeste.csv')
df_Rsuldeste = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/TRABALHO/historico_regiao_sudeste.csv')

# Converter a coluna 'Unnamed: 0' para 'Ano'
df_Rnorte.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)
df_Rnordeste.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)
df_Rsul.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)
df_RcentroOeste.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)
df_Rsuldeste.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)

# Converter a coluna 'Ano' para numérico
df_Rnorte['Ano'] = pd.to_numeric(df_Rnorte['Ano'], errors='coerce')
df_Rnordeste['Ano'] = pd.to_numeric(df_Rnordeste['Ano'], errors='coerce')
df_Rsul['Ano'] = pd.to_numeric(df_Rsul['Ano'], errors='coerce')
df_RcentroOeste['Ano'] = pd.to_numeric(df_RcentroOeste['Ano'], errors='coerce')
df_Rsuldeste['Ano'] = pd.to_numeric(df_Rsuldeste['Ano'], errors='coerce')

# Filtrar os últimos 5 anos em cada dataset
df_Rnorte_ultimos_5 = df_Rnorte[df_Rnorte['Ano'] >= 2019];
df_Rnordeste_ultimos_5 = df_Rnordeste[df_Rnordeste['Ano'] >= 2019];
df_Rsul_ultimos_5 = df_Rsul[df_Rsul['Ano'] >= 2019];
df_RcentroOeste_ultimos_5 = df_RcentroOeste[df_RcentroOeste['Ano'] >= 2019];
df_Rsuldeste_ultimos_5 = df_Rsuldeste[df_Rsuldeste['Ano'] >= 2019];

# Adicionar a coluna 'Regiao' para cada dataset
df_Rnorte_ultimos_5['Regiao'] = 'Norte'
df_Rnordeste_ultimos_5['Regiao'] = 'Nordeste'
df_Rsul_ultimos_5['Regiao'] = 'Sul'
df_RcentroOeste_ultimos_5['Regiao'] = 'Centro-Oeste'
df_Rsuldeste_ultimos_5['Regiao'] = 'Sudeste'

# Combinar os datasets em um único dataframe
df_completo = pd.concat([df_Rnorte_ultimos_5, df_Rnordeste_ultimos_5, df_Rsul_ultimos_5, df_RcentroOeste_ultimos_5, df_Rsuldeste_ultimos_5])

# Agrupar por Ano e Região, somando o total de queimadas
df_agrupado = df_completo.groupby(['Ano', 'Regiao']).sum().reset_index()

# Configurar o tamanho do gráfico
plt.figure(figsize=(12, 8))

# Criar o gráfico de linha com seaborn
sns.lineplot(data=df_agrupado, x='Ano', y='Total', hue='Regiao', marker='o')

# Adicionar títulos e rótulos
plt.title('Total de Queimadas por Região nos Últimos 5 Anos')
plt.xlabel('Ano')
plt.ylabel('Total de Queimadas')

# Exibir o gráfico
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.legend(title='Região')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os datasets
df_ce = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_ceara.csv')
df_ba = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_bahia.csv')
df_ma = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_maranhao.csv')
df_pb = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_paraiba.csv')
df_pe = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_pernambuco.csv')
df_pi = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_piaui.csv')
df_rn = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_rio_grande_do_norte.csv')
df_se = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_sergipe.csv')

for df in [df_ce, df_ba, df_ma, df_pb, df_pe, df_pi, df_rn, df_se]:
    df.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)

# Converter a coluna 'Ano' para numérico
for df in [df_ce, df_ba, df_ma, df_pb, df_pe, df_pi, df_rn, df_se]:
    df['Ano'] = pd.to_numeric(df['Ano'], errors='coerce')

# Filtrar os últimos 4 anos em cada dataset
ano_atual = 2023  # Ajuste o ano conforme necessário
df_ce_ultimos_4 = df_ce[df_ce['Ano'] >= (ano_atual - 4)]
df_ba_ultimos_4 = df_ba[df_ba['Ano'] >= (ano_atual - 4)]
df_ma_ultimos_4 = df_ma[df_ma['Ano'] >= (ano_atual - 4)]
df_pb_ultimos_4 = df_pb[df_pb['Ano'] >= (ano_atual - 4)]
df_pe_ultimos_4 = df_pe[df_pe['Ano'] >= (ano_atual - 4)]
df_pi_ultimos_4 = df_pi[df_pi['Ano'] >= (ano_atual - 4)]
df_rn_ultimos_4 = df_rn[df_rn['Ano'] >= (ano_atual - 4)]
df_se_ultimos_4 = df_se[df_se['Ano'] >= (ano_atual - 4)]

# Adicionar a coluna 'Estado' para cada dataset
df_ce_ultimos_4['Estado'] = 'Ceará'
df_ba_ultimos_4['Estado'] = 'Bahia'
df_ma_ultimos_4['Estado'] = 'Maranhão'
df_pb_ultimos_4['Estado'] = 'Paraíba'
df_pe_ultimos_4['Estado'] = 'Pernambuco'
df_pi_ultimos_4['Estado'] = 'Piauí'
df_rn_ultimos_4['Estado'] = 'Rio Grande do Norte'
df_se_ultimos_4['Estado'] = 'Sergipe'

# Combinar os datasets em um único dataframe
df_completo = pd.concat([df_ce_ultimos_4, df_ba_ultimos_4, df_ma_ultimos_4,
                         df_pb_ultimos_4, df_pe_ultimos_4, df_pi_ultimos_4,
                         df_rn_ultimos_4, df_se_ultimos_4], ignore_index=True)

# Agrupar os dados para calcular o total de queimadas por estado e ano
df_grouped = df_completo.groupby(['Estado', 'Ano'])['Total'].sum().reset_index()

# Ordenar os dados por ano e total de queimadas do maior para o menor
df_grouped = df_grouped.sort_values(by=['Ano', 'Total'], ascending=[True, False])

# Configurar o tamanho do gráfico
plt.figure(figsize=(12, 8))

# Criar o gráfico de barras com seaborn
sns.barplot(x='Ano', y='Total', hue='Estado', data=df_grouped)

# Adicionar títulos e rótulos
plt.title('Total de Queimadas por Estado nos Últimos 4 Anos')
plt.xlabel('Ano')
plt.ylabel('Total de Queimadas')

# Exibir o gráfico
plt.legend(title='Estado', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
df_ce = pd.read_csv('/content/drive/MyDrive/PESSOAL/ESTUDOS/ESTACIO/CURSOS PYTHON/DATASETS/ESTADO - NORDESTE/historico_estado_ceara.csv')

# Renomear a coluna 'Unnamed: 0' para 'Ano'
df_ce.rename(columns={'Unnamed: 0': 'Ano'}, inplace=True)

# Transformar os dados para o formato longo
df_long = pd.melt(df_ce, id_vars=['Ano'],
                  value_vars=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio',
                              'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro',
                              'Novembro', 'Dezembro'],
                  var_name='Mes', value_name='Valor')

# Criar um dicionário para mapear meses em português para inglês
meses_portugues = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'Março': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}

# Substituir os nomes dos meses no DataFrame
df_long['Mes'] = df_long['Mes'].map(meses_portugues)

# Limpar dados inválidos
df_long = df_long.dropna(subset=['Mes', 'Valor'])

# Converter as colunas 'Valor' e 'Ano' para numérico
df_long['Valor'] = pd.to_numeric(df_long['Valor'], errors='coerce')
df_long['Ano'] = pd.to_numeric(df_long['Ano'], errors='coerce')

# Filtrar para mostrar apenas os dados válidos após a conversão
df_long = df_long.dropna(subset=['Ano'])

# Agrupar os dados por 'Ano' e calcular a soma total por ano
soma_por_ano = df_long.groupby('Ano')['Valor'].sum().reset_index()

# Encontrar o valor máximo por ano
soma_por_ano['Maximo'] = soma_por_ano['Valor']

# Criar um gráfico de barras com os valores máximos por ano
plt.figure(figsize=(12, 6))
sns.barplot(data=soma_por_ano, x='Ano', y='Maximo', palette='viridis')

# Adicionar títulos e rótulos
plt.title('Análise do Total de Queimadas por Ano no Ceará')
plt.xlabel('Ano')
plt.ylabel('Soma Total de Valores')

# Exibir o gráfico
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
